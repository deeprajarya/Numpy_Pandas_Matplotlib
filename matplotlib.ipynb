```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
list1=[1,2,3,4,5,6]
list2=[11,21,23,12,14,15]

print(list1)
print(list2)



a1=np.arange(0,16)
a2= np.arange(10,26)
print("a1 =", a1)
print("a2 =",a2)
```

```python
plt.scatter(a1,a2, c="r",)
plt.xlabel("X -Axis")
plt.ylabel("Y-Axis")
plt.title("Graph")
plt.savefig("Img.png") 
```

```python
plt.scatter(list1,list2, c="b")
plt.xlabel("X -Axis")
plt.ylabel("Y-Axis")
plt.title("Graph")
plt.savefig("Img2.png")               # To save the figure in the particular directories.
```

```python
# plotting Graph
# plt.plot is used for line plot

plt.plot(a1,a2)
plt.plot(list1,list2)
```

```python
x=a1*a2
plt.plot(x)
```

```python
plt.plot(x,"r*")
```

```python
# line-style

plt.plot(x,a1,"b*",linestyle="dashed")
```

```python
# subplots

plt.subplot(2,2,1)
plt.plot(x,a1,'r--')
plt.subplot(2,2,2)
plt.plot(x,a2,'b--')
plt.subplot(2,2,3)
plt.plot(x,a1,'b--')
plt.subplot(2,2,4)
plt.plot(x,a2,'b--')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# Bar Plot

plt.bar(a1,a2)
```

```python
plt.bar(list1,list2,color="g")
```
